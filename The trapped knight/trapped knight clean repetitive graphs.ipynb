```python
class spiralboard:
    '''A class to create a spiral board where bigger numbers will be spiralling out'''
    
    def __init__(self, len_of_square_grid, direction):
        self.environment = pd.DataFrame(np.zeros((len_of_square_grid, len_of_square_grid)))
        self.direction = direction
        starting_square = (len_of_square_grid)//2
        self.x = starting_square
        self.y = starting_square
        self.environment[self.x][self.y] = 1
        self.count = 2
    
    def move(self):          
        if self.direction == 'East':
            if self.environment[self.x][self.y - 1] == 0:
                self.y = (self.y - 1)
                self.direction = 'South'
            else:
                self.x = self.x + 1
            self.environment[self.x][self.y] = self.count
            self.count = self.count + 1
        
        elif self.direction == 'North':
            if self.environment[self.x + 1][self.y] == 0:
                self.x = (self.x + 1)
                self.direction = 'East'
            else:
                self.y = self.y + 1
            self.environment[self.x][self.y] = self.count
            self.count = self.count + 1
                
        elif self.direction == 'West':
            if self.environment[self.x][self.y + 1] == 0:
                self.y = (self.y + 1)
                self.direction = 'North'
            else:
                self.x = self.x - 1
            self.environment[self.x][self.y] = self.count
            self.count = self.count + 1
                
        elif self.direction == 'South':
            if self.environment[self.x - 1][self.y] == 0:
                self.x = (self.x - 1)
                self.direction = 'West'
            else:
                self.y = self.y - 1
            self.environment[self.x][self.y] = self.count
            self.count = self.count + 1
                                     
test = spiralboard(101, direction = 'East')

len_of_square_grid = 101
for i in range(len_of_square_grid*len_of_square_grid - 1):
    test.move()

env = test.environment

```

```python
class knight:
    
    '''class to track the moves of knight'''
    
    def __init__(self, envi,pattern,len_of_square_grid):   
        starting_square = (len_of_square_grid)//2
        self.small_step = min(pattern)
        print('Type of small step is',self.small_step)
        self.big_step = max(pattern)
        self.envi = envi
        self.x = starting_square
        self.y = starting_square
        self.big_num = (envi).shape[0]*((envi).shape[0] + 1)
        self.prev = 0 
        self.xmin = envi.shape[0]
        self.xmax = 0
        self.ymin = envi.shape[0]
        self.ymax = 0
        
    def moves(self):
        
        global continu
        
        allMoves = []
        try:
            move0 = self.envi[self.x + self.small_step][self.y + self.big_step]
        except:
            move0 = self.big_num
        try:
            move1 = self.envi[self.x + self.small_step][self.y - self.big_step]
        except:
            move1 = self.big_num    
        try:
            move2 = self.envi[self.x + self.big_step][self.y + self.small_step]
        except:
            move2 = self.big_num
        try:
            move3 = self.envi[self.x + self.big_step][self.y - self.small_step]
        except:
            move3 = self.big_num
        try:
            move4 = self.envi[self.x - self.small_step][self.y + self.big_step]
        except:
            move4 = self.big_num    
        try:
            move5 = self.envi[self.x - self.small_step][self.y - self.big_step]
        except:
            move5 = self.big_num
        try:
            move6 = self.envi[self.x - self.big_step][self.y + self.small_step]
        except:
            move6 = self.big_num
        try:
            move7 = self.envi[self.x - self.big_step][self.y - self.small_step]
        except:
            move7 = self.big_num
        
        allMoves.extend([move0, move1, move2, move3, move4, move5, move6, move7])
        
        self.prev = self.envi[self.x][self.y]
        self.envi[self.x][self.y] = self.big_num
        self.xmax = max(self.xmax,self.x)
        self.xmin = min(self.xmin,self.x)
        self.ymax = max(self.ymax,self.y)
        self.ymin = min(self.ymin,self.y)
        
        if min(allMoves) >= self.big_num:
            continu  = False
            print("knight trapped at the square" ,self.prev, 'that is for following respective x and y coordinates',self.x, self.y)
        
        bestmove = allMoves.index(min(allMoves))
        
        if bestmove == 0 and ((self.x + self.small_step) < (len_of_square_grid-1)) and ((self.y + self.big_step)<(len_of_square_grid-1)):
            self.x = (self.x + self.small_step)
            self.y = (self.y + self.big_step)
        elif bestmove == 1 and ((self.x + self.small_step) < (len_of_square_grid-1)) and ((self.y - self.big_step)>0):
            self.x = (self.x + self.small_step)
            self.y = (self.y - self.big_step)
        elif bestmove == 2 and ((self.x + self.big_step) < (len_of_square_grid-1)) and ((self.y + self.small_step)<(len_of_square_grid-1)):
            self.x = (self.x + self.big_step)
            self.y = (self.y + self.small_step)
        elif bestmove == 3 and ((self.x + self.big_step) < (len_of_square_grid-1)) and ((self.y - self.small_step)>0):
            self.x = (self.x + self.big_step)
            self.y = (self.y - self.small_step)
        elif bestmove == 4 and ((self.x - self.small_step) > 0) and ((self.y + self.big_step)<(len_of_square_grid-1)):
            self.x = (self.x - self.small_step)
            self.y = (self.y + self.big_step)
        elif bestmove == 5 and ((self.x - self.small_step) > 0) and ((self.y - self.big_step)>0):
            self.x = (self.x -self.small_step)
            self.y = (self.y - self.big_step)
        elif bestmove == 6 and ((self.x - self.big_step) > 0) and ((self.y + self.small_step)<(len_of_square_grid-1)):
            self.x = (self.x - self.big_step)
            self.y = (self.y + self.small_step)
        elif bestmove == 7 and ((self.x - self.big_step) > 0) and ((self.y - self.small_step)>0):
            self.x = (self.x - self.big_step)
            self.y = (self.y - self.small_step)
        else:
            continu = False
            print("knight trapped at the square" ,self.prev, 'that is for following respective x and y coordinates',self.x, self.y)
            print('knight is trapped because of insufficient squares, you can try again by increasing the number of square if you truly want to know when will it be trapped')


# create member of knight class
len_of_square_grid = int(input('please input the size of the chess board\n'))
center = (len_of_square_grid)//2
test = spiralboard(len_of_square_grid, direction = 'East')
for i in range(len_of_square_grid*(len_of_square_grid - 1)):
    test.move()
env= test.environment
print('''printing the central part of the board, just to crosscheck''')
print(env[center-4:center+5].T[center-4:center+5])
pattern = list(map(int, input('''provide two space seperated numbers according to which the knight moves on chess board,
Ex:input of 1 2 will make the knight move conventionally i.e 1 step in one direction and 2 steps in direction perperndicular to it \n''').split()))
if pattern:
    if min(pattern)<1 or (len(pattern)!=2):
        pattern = [1,2]
        print('Using conventional rules for the movement of knight as you have not provided valid integers')
else:
    pattern = [1,2]
    print('Using conventional rules for the movement of knight as you have not provided valid integers')

    

horse = knight(env,pattern,len_of_square_grid)
continu= 1

# list to store visited coordinates
history = []

#keep on moving until it gets stuck
while continu:
    history.append([horse.x, horse.y, horse.prev])
    '''uncomment the next line if you want to know the running process of the program'''
#     print('processed the square',env[horse.x][horse.y])
    horse.moves()   
xmax = horse.xmax
xmin = horse.xmin
ymax = horse.ymax
ymin = horse.ymin
print('sadly it was able to make only',len(history),'steps')
trapped_square = history[-1]

# get X's and Y's
x = [history[i][0] for i in range(len(history))]
y = [history[i][1] for i in range(len(history))]
colors= np.array([history[i][2] for i in range(len(history))])
colors1 = np.arange(1,trapped_square[2]+1)

#normalizing the colors so that they can be mapped with colormap
p = colors/trapped_square[2]   #this is the numbers of squares it has covered, so it should be mostly square type diverging

points = np.array([x,y]).T.reshape(-1,1,2)
print(points.shape)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
print(segments.shape)  

lc = col.LineCollection(segments, cmap=plt.get_cmap('hsv'))#RdYlBu,jet,nipy_spectral,hsv
lc.set_array(p)
lc.set_linewidth(1.5)

fig1 = plt.figure(figsize=(15, 15), num = 1)
plt.gca().add_collection(lc)
plt.title('the number of the square')

'''plotting only the region of interest'''
plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

'''plotting the struck point'''
plt.plot(x[-1], y[-1],marker ='o', color='green', markersize = 6)
plt.show()

#normalizing the colors so that they can be mapped with colormap
p1 = colors1/trapped_square[2] #this is the order of squares it has covered, so it will give you the actual path of the knight

lc1 = col.LineCollection(segments, cmap=plt.get_cmap('hsv'))#RdYlBu,jet,nipy_spectral,hsv
lc1.set_array(p1)
lc1.set_linewidth(1.5)

'''plotting only the region of interest'''
fig2 = plt.figure(figsize=(15, 15), num = 2)
plt.gca().add_collection(lc1)
plt.title('the number of the move')

plt.xlim(xmin,xmax)
plt.ylim(ymin,ymax)

'''plotting the struck point'''
plt.plot(x[-1], y[-1],marker ='o', color='green', markersize = 6)
plt.show()
fig2.savefig('trial.pdf', bbox_inches='tight')
patterns = []
```
