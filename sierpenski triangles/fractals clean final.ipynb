```python
import numpy as np
import math
import random
from numpy import array as arr
import matplotlib.pyplot as plt
from matplotlib import interactive
```

```python
num_points = np.logspace(2,7,6)
```

```python
fig_num = 1
edgelen = 100
fixed_vertices = {'v1':  arr([0,0]), 'v2' : arr([edgelen,0]), 'v3' : arr([edgelen/2,(math.sqrt(3)*edgelen)/2])}
# curr_fixed_v = v1
for num_pts in num_points:
#     print(num_pts)
#     pts = [np.zeros(2) for i in range(int(num_pts))]
    pts = np.zeros(shape = (int(num_pts),2))
#     pts [1] = arr([edgelen/2,edgelen/2])
    pts[1] = arr([edgelen/2,edgelen/2])
    for i in range(1,int(num_pts)):
        curr_fixed_v = fixed_vertices['v'+str(random.randint(1,3))]
        pts[i] = (pts[i-1] + curr_fixed_v)/2
    figure = plt.figure(figsize= (25,25), num =fig_num)
    plt.scatter(np.array(pts).T[0],np.array(pts).T[1])
    title = 'sierpinski triangles _ num pts ' + str(int(num_pts))
    plt.title(title)
    fig_num+=1
    figure.show()
    interactive(True)
#     figure.savefig(title+'.png')
interactive(False)
```

```python
num_points = np.logspace(2,7,6)
```

```python
fig_num = 1
edgelen = 100
fixed_vertices = {'v1':  arr([0,0]), 'v2' : arr([edgelen,0]), 'v3' : arr([edgelen/2,(math.sqrt(3)*edgelen)/2])}
for num_pts in num_points:
    print(num_pts)
#     pts = [np.zeros(2) for i in range(int(num_pts))]
    pts = np.zeros(shape = (int(num_pts),2))
#     pts [1] = arr([edgelen/2,edgelen/2])
    pts[1] = arr([edgelen/2,edgelen/2])
    for i in range(1,int(num_pts)):
        curr_fixed_v = fixed_vertices['v'+str(random.randint(1,3))]
        pts[i] = (pts[i-1] + curr_fixed_v)/2
    figure = plt.figure(figsize= (25,25), num =fig_num)
#     plt.scatter(np.array(pts).T[0],np.array(pts).T[1])
    title = 'sierpinski triangles _ num pts ' + str(int(num_pts))
    scatter = hv.Scatter(data = np.array(pts))
    scatter.opts(title='Sierpenski Triangles {}'.format("n_points = " + str(int(num_pts))),width=600, height=600, show_grid=True)
    hv.save(scatter, 'images/{}.html'.format(title), fmt='html')
    scatter
```
